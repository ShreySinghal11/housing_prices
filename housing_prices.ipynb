```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
 ## for same values of null values for all inputs
    
data.dropna(inplace=True)
```

```python
data.info()
```

```python
pip install scikit-learn
```

```python
 ## training the data
    
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'],axis=1)
y = data['median_house_value']
```

```python
y
```

```python
#train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
train_data=x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(16,8))
```

```python
#correlation of matrix
plt.figure(figsize=(16,8))
sns.heatmap(train_data.corr(),annot=True, cmap="YlGnBu")
```

```python
train_data.ocean_proximity.value_counts()
```

```python
pd.get_dummies(train_data.ocean_proximity)
```

```python
#plotting in the graph according to the location

plt.figure(figsize=(16,8))
sns.scatterplot(x="latitude",y="longitude",data=train_data, hue="median_house_value", palette="coolwarm")
```

```python
# We see that the houses close to the sea are more expensive than the ones awat from it
# the prices is shown by the colored dots- the more expensive ones are shown by red and the less expensive ones by blue
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms']/ train_data['households']
```
